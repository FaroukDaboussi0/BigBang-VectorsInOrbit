```python
# Calculation of the new EMI 
import numpy as np

def calculate_new_emi(df):
    """
    Calculates the EMI for the requested loan based on amount, rate, and tenure.
    """
    P = df['loan_amount_requested']
    annual_rate = df['interest_rate_offered']
    n = df['loan_tenure_months']
    
    # 1. Convert annual interest rate to monthly decimal
    # Example: 12% annual -> 0.12 / 12 = 0.01 monthly
    r = annual_rate / (12 * 100)
    
    # 2. Apply EMI Formula: [P x r x (1+r)^n] / [(1+r)^n - 1]
    # We use np.where to handle cases where rate might be 0 (though rare) to avoid division by zero
    emi = np.where(
        r > 0,
        (P * r * (1 + r)**n) / ((1 + r)**n - 1),
        P / n # If interest is 0, it's just principal divided by months
    )
    return emi

# --- Applying it to your dataframe ---

# Step 1: Calculate the New EMI for the requested loan
df['estimated_new_emi'] = calculate_new_emi(df)

# Step 2: Calculate the Total Debt Burden (Existing + New)
df['total_monthly_debt'] = df['existing_emis_monthly'] + df['estimated_new_emi']

# Step 3: Calculate the Installment to Income Ratio (DTI 2.0)
df['installment_to_income_ratio'] = df['total_monthly_debt'] / df['monthly_income']

# Optional: Description for your documentation
feature_explanation = {
    'estimated_new_emi': 'The calculated monthly repayment for the loan currently being applied for.',
    'installment_to_income_ratio': 'Total debt obligations (old + new) divided by monthly income. '
                                   'Values > 0.5 usually indicate high credit risk (Over-leveraged).'
}

print("New feature 'installment_to_income_ratio' calculated successfully.")
print(df[['loan_amount_requested', 'estimated_new_emi', 'installment_to_income_ratio']].head())
```

## 3- Feature and vector preparation for the risk vector 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# ==========================================
# 1. AGGREGATE TRANSACTIONS (Behavioral Data)
# ==========================================
# Define our Category Lists based on your distinct values
essential_categories = ['Utilities', 'Healthcare', 'Groceries', 'Education', 'Fuel']
lifestyle_categories = ['Dining', 'Travel', 'Entertainment', 'Online Shopping', 'Electronics']
fixed_transaction_types = ['Bill Payment', 'EMI Payment']

# Assuming train_transaction_df has 'transaction_date' converted to datetime
train_transaction_df['transaction_date'] = pd.to_datetime(train_transaction_df['transaction_date'])
train_transaction_df['month_year'] = train_transaction_df['transaction_date'].dt.to_period('M')

agg_tx = train_transaction_df.groupby('customer_id').agg(
    avg_monthly_balance=('account_balance_after_transaction', 'mean'),
    balance_volatility=('account_balance_after_transaction', 'std'),
    min_balance_reached=('account_balance_after_transaction', 'min'),
    max_transaction_value=('transaction_amount', 'max'),
    avg_transaction_value=('transaction_amount', 'mean'),
    total_spent_ever=('transaction_amount', 'sum'),
    transaction_frequency=('transaction_id', 'count'),
    international_tx_indicator=('is_international_transaction', 'sum'),
    device_diversity_score=('device_used', 'nunique'),
    ip_stability_ratio=('ip_address', 'nunique'),
    # For essential/lifestyle ratios (example categories)
    # 1. ESSENTIAL SPEND (Fixed Obligations + Survival)
    # Logic: Sum amount if Type is Bill/EMI OR Category is Utilities/Healthcare/Groceries/Education/Fuel
    essential_spend=('transaction_amount', lambda x: x[
        (train_transaction_df.loc[x.index, 'transaction_type'].isin(fixed_transaction_types)) | 
        (train_transaction_df.loc[x.index, 'merchant_category'].isin(essential_categories))
    ].sum()),  

    # 2. LIFESTYLE SPEND (Discretionary/Flexible)
    # Logic: Sum amount if Category is Dining/Travel/Entertainment/Shopping/Electronics
    lifestyle_spend=('transaction_amount', lambda x: x[
        train_transaction_df.loc[x.index, 'merchant_category'].isin(lifestyle_categories)
    ].sum()),  

    # 3. FINANCIAL SERVICES / DEBT (Specific check)
    debt_related_spend=('transaction_amount', lambda x: x[
        train_transaction_df.loc[x.index, 'merchant_category'] == 'Financial Services'
    ].sum()),

    # 4. CASH LIQUIDITY (ATM behavior)
    cash_withdrawal_total=('transaction_amount', lambda x: x[
        (train_transaction_df.loc[x.index, 'transaction_type'] == 'ATM Withdrawal') |
        (train_transaction_df.loc[x.index, 'merchant_category'] == 'Cash Withdrawal')
    ].sum())
)

# Calculate transaction-specific engineered features
total_months = train_transaction_df.groupby('customer_id')['month_year'].nunique()
agg_tx['total_monthly_burn_rate'] = agg_tx['total_spent_ever'] / total_months
agg_tx['essential_spending_ratio'] = agg_tx['essential_spend'] / agg_tx['total_spent_ever']
agg_tx['lifestyle_spending_ratio'] = agg_tx['lifestyle_spend'] / agg_tx['total_spent_ever']
agg_tx['transaction_frequency'] = agg_tx['transaction_frequency'] / total_months
agg_tx['ip_stability_ratio'] = agg_tx['ip_stability_ratio'] / agg_tx['transaction_frequency']
agg_tx['debt_indicator_ratio'] = agg_tx['debt_related_spend'] / (agg_tx['total_spent_ever'] + 1)
agg_tx['cash_dependence_ratio'] = agg_tx['cash_withdrawal_total'] / (agg_tx['total_spent_ever'] + 1)

# ==========================================
# 2. MERGE DATASETS
# ==========================================
risk_master_df = train_loan_df.merge(agg_tx, on='customer_id', how='left').fillna(0)

# ==========================================
# 3. LOAN APPLICATION ENGINEERING
# ==========================================
# Estimated New EMI (Rough approx: Principal / Tenure * Interest factor)
risk_master_df['estimated_new_emi'] = (risk_master_df['loan_amount_requested'] / risk_master_df['loan_tenure_months']) * 1.1

risk_master_df['loan_to_income_ratio'] = risk_master_df['loan_amount_requested'] / (risk_master_df['monthly_income'] * 12)
risk_master_df['installment_to_income_ratio'] = (risk_master_df['existing_emis_monthly'] + risk_master_df['estimated_new_emi']) / risk_master_df['monthly_income']
risk_master_df['disposable_income'] = risk_master_df['monthly_income'] - risk_master_df['existing_emis_monthly']
risk_master_df['age_at_loan_end'] = risk_master_df['applicant_age'] + (risk_master_df['loan_tenure_months'] / 12)
risk_master_df['income_per_dependent'] = risk_master_df['monthly_income'] / (risk_master_df['number_of_dependents'] + 1)

# ==========================================
# 4. CROSS-DATASET RATIOS
# ==========================================
risk_master_df['cash_flow_coverage_ratio'] = risk_master_df['avg_monthly_balance'] / (risk_master_df['total_monthly_burn_rate'] + 1)
risk_master_df['payment_to_income_reality_check'] = risk_master_df['essential_spend'] / (risk_master_df['monthly_income'] + 1)
risk_master_df['income_stability_proxy'] = risk_master_df['balance_volatility'] / (risk_master_df['avg_monthly_balance'] + 1)

# ==========================================
# 5. CATEGORICAL ENCODING (For Vectors)
# ==========================================
# Maps categories to numbers so the vector distance makes sense
risk_master_df['gender_val'] = risk_master_df['gender'].map({'Male': 0, 'Female': 1, 'Other': 0.5})
risk_master_df['property_val'] = risk_master_df['property_ownership_status'].map({'Rented': 0, 'Jointly Owned': 0.5, 'Owned': 1})
risk_master_df['cibil_category_val'] = pd.cut(risk_master_df['cibil_score'], bins=[0, 600, 700, 800, 900], labels=[0, 0.33, 0.66, 1]).astype(float)

# ==========================================
# 5. CATEGORICAL ENCODING (For Vectors)
# ==========================================
# Count how many times this customer has appeared in the loan dataset
risk_master_df['previous_application_count'] = risk_master_df.groupby('customer_id')['application_id'].transform('count')

# ==========================================
# 6. FINAL VECTOR FEATURE LIST (30+ Dimensions)
# ==========================================
vector_features = [
    # Application Raw
    'applicant_age', 'number_of_dependents', 'monthly_income', 'cibil_score', 
    'existing_emis_monthly', 'debt_to_income_ratio', 'loan_amount_requested', 
    'loan_tenure_months', 'interest_rate_offered',
    
    # Application Engineered
    'loan_to_income_ratio', 'installment_to_income_ratio', 'disposable_income', 
    'age_at_loan_end', 'income_per_dependent', 'gender_val', 'property_val', 'cibil_category_val',
    
    # Transactional Behavioral
    'avg_monthly_balance', 'balance_volatility', 'min_balance_reached', 
    'max_transaction_value', 'avg_transaction_value', 'total_monthly_burn_rate', 
    'essential_spending_ratio', 'lifestyle_spending_ratio', 'transaction_frequency', 
    'international_tx_indicator', 'device_diversity_score', 'ip_stability_ratio',
    'debt_indicator_ratio', 'cash_dependence_ratio',
    
    # Cross-Dataset
    'cash_flow_coverage_ratio', 'payment_to_income_reality_check', 'income_stability_proxy',
    # 'previous_application_count'
    'previous_application_count'
]

# ==========================================
# 7. SCALING & VECTORIZATION
# ==========================================
scaler = MinMaxScaler()
# Clean any infinities from division by zero
risk_master_df[vector_features] = risk_master_df[vector_features].replace([np.inf, -np.inf], 0)

risk_master_df['risk_vector'] = scaler.fit_transform(risk_master_df[vector_features]).tolist()

print(f"Vectorization complete. Vector size: {len(vector_features)} dimensions.")
```

```python
import json
from groq import Groq

# 1. Initialize Groq Client
groq_client = Groq(api_key="API_KEY_HERE")  # Replace with your Groq API Key

def orchestrate_decision(new_customer_raw_data, scaler, qdrant_client):
    """
    Full Orchestrator: Raw Data -> Vector -> Qdrant Search -> Groq Decision
    """
    
    # --- STEP 3.1: Transform New Customer to Risk Vector ---
    # NOTE: You must use the SAME scaler and feature logic from Phase 1
    # feature_list = [applicant_age, cibil_score, loan_to_income_ratio, etc...]
    try:
        # Pull only the 35 features needed for the vector
        feature_values = [new_customer_raw_data[feat] for feat in vector_features]

        # Scale the features (Scaler expects a 2D array, so we use [feature_values])
        new_vector = scaler.transform([feature_values])[0].tolist()
    except KeyError as e:
        return f"Error: Missing feature in input data: {e}"

    # --- STEP 3.2: Semantic Search in Qdrant ---
    search_result = qdrant_client.query_points(
        collection_name="credit_decision_memory",
        query=new_vector,
        limit=3, # Get the 3 closest Financial Twins
        with_payload=True
    )

    # --- STEP 3.3: Format Data for Groq ---
    # Prepare the "Memory" context for the LLM
    similar_cases_context = ""
    for i, hit in enumerate(search_result.points):
        p = hit.payload
        # Using the payload columns we defined in Phase 2
        similar_cases_context += f"""
        Twin #{i+1} (Similarity Score: {hit.score:.4f}):
        - Status: {p.get('loan_status', 'N/A')}
        - CIBIL: {p.get('cibil_score', 'N/A')}
        - Income: ${p.get('monthly_income', 'N/A')}
        - Cash Flow Coverage: {p.get('cash_flow_coverage_ratio', 'N/A')}
        - Lifestyle Ratio: {p.get('lifestyle_spending_ratio', 'N/A')}
        ---
        """

    # --- STEP 3.4: Construct the Prompt ---
    # We use the payload_columns list to select what Grok sees about the new app
    new_app_summary = {col: new_customer_raw_data.get(col, "N/A") for col in payload_columns}

    prompt = f"""
    ROLE: Senior Credit Risk Underwriter (AI)
    TASK: Decide on a new Loan Application by comparing it to 'Historical Memory'.

    NEW APPLICANT DATA:
    {json.dumps(new_app_summary, indent=2)}

    HISTORICAL LOOK-ALIKES FROM MEMORY:
    {similar_cases_context}

    ANALYSIS GUIDELINES:
    - If the Twins with similar 'Cash Flow Coverage' and 'DTI' defaulted (Risk), be cautious.
    - Look at the 'income_stability_proxy'. If it's worse than the Twins, note that.
    - Consider the 'purpose_of_loan' and its alignment with spending habits.

    OUTPUT FORMAT:
    1. FINAL_STATUS: [APPROVED or REJECTED]
    2. CONFIDENCE_SCORE: [0-100%]
    3. UNDERWRITER_EXPLANATION: [3-5 sentences explaining the decision based on similarity to history]
    """

    # --- STEP 3.5: Call Groq API ---
    completion = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile", # Or 'mixtral-8x7b-32768'
        messages=[
            {"role": "system", "content": "You are a precise banking credit decision engine."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.1, # Keep it low for consistent financial decisions
        max_tokens=500
    )

    return completion.choices[0].message.content

# --- TEST THE ORCHESTRATOR ---
#new_app = {... raw dictionary of features ...}
#decision = orchestrate_decision(new_app, scaler, client)
#print(decision)
```

```python
from qdrant_client import QdrantClient
from qdrant_client.http import models
import numpy as np

# 1. Initialize Client
client = QdrantClient(
        url=ENDPOINT, 
        api_key=API_KEY,
    )

# 2. Define Collection Parameters
COLLECTION_FRAUD = "fraud_anomaly_memory"
VECTOR_SIZE = len(fraud_vector_features) # This will be 25 based on your list

# Create Collection
client.recreate_collection(
    collection_name=COLLECTION_FRAUD,
    vectors_config=models.VectorParams(
        size=len(fraud_vector_features), 
        distance=models.Distance.COSINE 
    ),
)

# Prepare Fraud-Specific Payload
# These columns help the LLM identify the type of fraud occurring
fraud_payload_columns = [
    # 1. THE OUTCOME (The "Truth" from History)
    'customer_id',
    'fraud_flag',                # 0 or 1
    'fraud_type',                # e.g., 'Synthetic Identity', 'Income Misrepresentation'
    'loan_status',               # e.g., 'Fraudulent - Detected', 'Declined'

    # 2. THE RAW EVIDENCE (What the applicant claimed vs. Reality)
    'monthly_income',            # Raw Claimed Income
    'loan_amount_requested',     # Raw Loan Request
    'avg_tx_amount',             # Raw Spending Reality
    'sum_tx_amount',             # Total money moved in 30 days
    'total_tx_count',            # How active is this person?
    'employment_status',         # Raw Employment (Salaried, Student, etc.)

    # 3. THE IDENTITY SMOKING GUNS (Raw Network Data)
    'unique_ips',                # How many IPs used?
    'unique_devices',            # How many physical devices?
    'max_ip_sharing_score',      # How many OTHER people use this IP? (Crucial)
    'max_device_sharing_score',  # How many OTHER people use this Phone/PC?

    # 4. THE ENGINEERED ANOMALIES (The "Context" for the LLM)
    'income_validation_ratio',   # Reality Check: Income vs Spend
    'failed_ratio',              # Technical Check: Brute force/Testing indicator
    'midnight_app_flag',         # Behavioral Check: Suspicious timing
    'suspicious_notes_count',    # Content Check: Fraudulent keywords found
    'mule_indicator_ratio'       # Flow Check: Money funneling pattern
]

# 3. Batch Upsert (Processing 50,000 rows in chunks of 1000)
batch_size = 1000
for i in range(0, len(fraud_master_df), batch_size):
    batch_df = fraud_master_df.iloc[i:i+batch_size]
    
    points = []
    for idx, row in batch_df.iterrows():
        # Clean the payload: convert numpy types to python types for Qdrant compatibility
        payload = {col: row[col] for col in fraud_payload_columns}
        # Ensure numbers are JSON serializable
        for k, v in payload.items():
            if isinstance(v, (np.int64, np.float64)):
                payload[k] = v.item()
        
        points.append(
            models.PointStruct(
                id=idx, # Using the dataframe index as ID
                vector=row['fraud_vector'],
                payload=payload
            )
        )
    
    client.upsert(
        collection_name=COLLECTION_FRAUD,
        points=points
    )
    print(f"Uploaded batch {i // batch_size + 1} / {len(fraud_master_df) // batch_size + 1}")

print("--- PHASE 2 COMPLETE: Fraud Anomaly Memory is Live ---")
```

```python
def orchestrate_decision_v5(new_customer_data, risk_scaler, fraud_scaler, qdrant_client):
    # --- 1. VECTORIZATION (Keep the fixed DataFrame logic) ---
    risk_df = pd.DataFrame([new_customer_data])[vector_features]
    fraud_df = pd.DataFrame([new_customer_data])[fraud_vector_features]
    risk_vec = risk_scaler.transform(risk_df)[0].tolist()
    fraud_vec = fraud_scaler.transform(fraud_df)[0].tolist()

    # --- 2. DUAL RETRIEVAL ---
    risk_results = qdrant_client.query_points(collection_name="credit_decision_memory", query=risk_vec, limit=3)
    fraud_results = qdrant_client.query_points(collection_name="fraud_anomaly_memory", query=fraud_vec, limit=3)

    # --- 3. CONTEXT PREPARATION (Adding "Match Strength" Labels) ---
    fraud_context = ""
    max_fraud_score = 0
    for hit in fraud_results.points:
        p = hit.payload
        max_fraud_score = max(max_fraud_score, hit.score)
        # We give the LLM clear instructions on how to interpret the score
        strength = "CRITICAL MATCH" if hit.score > 0.94 else "SURE MATCH" if hit.score > 0.90 else "MILD SUSPICION"
        fraud_context += f"""
        Match ({strength} - Score: {hit.score:.2f}):
        - Historical Fraud Type: {p.get('fraud_type')}
        - Pattern: Income ${p.get('monthly_income')} vs Spend ${p.get('avg_tx_amount')}
        - Shared Device Score: {p.get('max_device_sharing_score')}
        ---"""

    risk_context = ""
    for hit in risk_results.points:
        p = hit.payload
        risk_context += f"- Twin (Sim: {hit.score:.2f}): Outcome: {p.get('loan_status')}, CIBIL: {p.get('cibil_score')}\n"

    # --- 4. THE CALIBRATED PROMPT ---
    prompt = f"""
    ROLE: Senior Credit Committee Judge.
    
    APPLICANT PROFILE:
    - CIBIL Score: {new_customer_data['cibil_score']}
    - Monthly Income: ${new_customer_data['monthly_income']}
    - Device Sharing: {new_customer_data['max_device_sharing_score']} (Note: 1 is normal, >2 is suspicious)

    FRAUD MEMORY (Historical Red Flags):
    {fraud_context}

    CREDIT MEMORY (Financial Look-alikes):
    {risk_context}

    DECISION RULES:
    1. If the top Fraud Match score is BELOW 0.90, do NOT reject for fraud unless there is a huge 'Income vs Spend' lie.
    2. If the Credit Memory Twins are 'APPROVED' and have a high similarity (>0.92), prioritize them.
    3. High CIBIL applicants (>750) often have shared devices (family). Don't penalize them unless the Fraud Match is >0.95.
    4. ONLY REJECT if you are certain. If in doubt, and the Credit Twins are good, APPROVE.

    OUTPUT FORMAT (STRICT):
    FINAL_STATUS: [APPROVED or REJECTED]
    CONFIDENCE_LEVEL: [0-100]%
    REASONING: [1 concise sentence explaining the balance between Risk and Fraud memory]
    """

    # --- 5. CALL GROQ ---
    completion = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "system", "content": "You are a balanced bank underwriter. You protect the bank from fraud but ensure good customers get their loans."},
                  {"role": "user", "content": prompt}],
        temperature=0.0 # Extreme precision
    )

    return completion.choices[0].message.content
```
